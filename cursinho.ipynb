1 - Abrindo um arquivo excel com o Pandas.

```python
import pandas as pd

arquivo = pd.read_excel("cursinho.xlsx")
display(arquivo)
```

2 - Gerando um gráfico com a porcentagem de alunos presentes por Matéria.   

```python
import matplotlib.pyplot as plt

alunos_pres = arquivo['Alunos presentes']
mats = arquivo['Matérias']

plt.title("Alunos presentes no cursinho por matéria")
plt.pie(alunos_pres, labels = mats, autopct='%1.2f%%')
plt.show()
```

3- Gerando um gráfico com a taxa de faltas dos alunos por dia da semana.

```python
dias = arquivo['Dias da semana']
faltas = arquivo['Alunos com falta']

plt.bar(dias, faltas)
plt.show()
```

Análise dos dados

Com base na análise destes 2 gráficos é possível concluir em primeiro lugar que, a maior presença dos alunos ocorre nas aulas de Português, Matemática e História.
Além disso, os alunos possuem uma tendência a faltar com mais frequência conforme o final de semana se aproxima, sendo sábado o pico de ausências.
